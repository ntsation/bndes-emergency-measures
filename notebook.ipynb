```python
import pandas as pd
from fetch_data import fetch_data
from process_data import process_data
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = fetch_data()
process_data(df)
```

#### Exploratory Data Analysis
##### Basic Statistics

```python
df.describe()
```

##### Distribution Plot

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['quantidade_ou_valor'], bins=30, kde=True)
plt.title('Distribution of Amounts/Values')
plt.xlabel('Amount/Value')
plt.ylabel('Frequency')
plt.show()
```

##### Top Categories

```python
top_categories = df['categoria'].value_counts().head(10)
top_categories.plot(kind='bar', figsize=(10, 6))
plt.title('Top 10 Categories by Total Amount/Value')
plt.xlabel('Category')
plt.ylabel('Total Amount/Value')
plt.show()
```
